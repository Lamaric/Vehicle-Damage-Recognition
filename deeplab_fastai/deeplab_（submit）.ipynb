```python
#读取类别
coco=COCO( '/data/code/test/instances_train2017.json')
cats = coco.loadCats(coco.getCatIds())
nms=[cat['name'] for cat in cats]
nms.insert(0,"void")
```

```python
name2id = {v:k for k,v in enumerate(nms)}
void_code = name2id['void']
#指标和Loss
def acc_camvid(input, target):
    target = target.squeeze(1)
    mask = target != void_code
    return (input.argmax(dim=1)[mask]==target[mask]).float().mean()
metrics=acc_camvid


def lovasz_loss(pred, labels):
    pred = F.softmax(pred,dim=1)
    loss = tensor(L.lovasz_softmax(pred,labels, ignore=255))
    return loss
```

```python
def split_layers(model:nn.Module) -> List[nn.Module]:
    groups = [[model.backbone, ]]
    groups += [[model.aspp, model.decoder]]
    return groups
```

# Test metrics

```python
def validate_self(model:nn.Module, dl:DataLoader,evalutor=None, loss_func:OptLossFunc=None,cb_handler:Optional[CallbackHandler]=None,pbar:Optional[PBar]=None, average=True, n_batch:Optional[int]=None):
    "Calculate `loss_func` of `model` on `dl` in evaluation mode."
    model.eval()
    evalutor.reset()
    with torch.no_grad():
        val_losses,nums = [],[]
        if cb_handler: cb_handler.set_dl(dl)
        for xb,yb in progress_bar(dl, parent=pbar, leave=(pbar is not None)):
            if cb_handler: xb, yb = cb_handler.on_batch_begin(xb, yb, train=False)
            val_loss = loss_batch(model, xb, yb, loss_func, cb_handler=cb_handler)
            val_losses.append(val_loss)
            evalutor.add_batch(val_loss[0].argmax(dim=1).unsqueeze(1).cpu().numpy(),val_loss[1].cpu().numpy())
```

```python
from Evaluator import Evaluator

evaluator = Evaluator(32)

validate_self(learn.model,learn.data.train_dl,evalutor=evaluator)

Acc = evaluator.Pixel_Accuracy()
Acc_class = evaluator.Pixel_Accuracy_Class()
mIoU = evaluator.Mean_Intersection_over_Union()
FWIoU = evaluator.Frequency_Weighted_Intersection_over_Union()
print('val/mIoU', mIoU)
print('val/Acc', Acc)
print('val/Acc_class', Acc_class)
print('val/fwIoU', FWIoU)
```
